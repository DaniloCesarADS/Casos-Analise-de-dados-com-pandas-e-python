```python
# importando a biblioteca Pandas
import pandas as pd
```

```python
# Fazendo a leitura das 5 planilhas excel
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
df1.head()
```

```python
# Unindo as 5 planilhas como uma só database
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
# Inserindo os 5 primeiros itens da database unificada
df.head() 
```

```python
# Inserindo os 5 ultimos itens da database unificada
df.tail()
```

```python
# Pegando uma amostra de dados aleatória na database unificada
df.sample(20)
```

```python
# Identificando o tipo de dado em cada coluna
df.dtypes
```

```python
# Como a coluna Loja ID não é utilizada para cálculos, então vamos mudá-la para o tipo Object
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Tratando valores faltantes
```

```python
# Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
# Substituindo valores nulos pelo valor da média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
# Verificando a média de vendas. Releia as 5 planilhas e concatene-as novamente para ver  a nova média, já que os valores nulos foram substituidos
# pela média de vendas
df["Vendas"].mean()
```

```python
df.sample(20)
```

```python
df.isnull().sum()
```

```python
# Substituindo valores nulos pelo valor 0
df["Vendas"].fillna(0, inplace=True)
```

```python
# Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
# Apagando as linhas com valores nulos com base em apenas 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
# Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

```python
# Criando novas colunas
```

```python
# Criando uma coluna adicional para o cálculo da receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
# Identificando a maior receita
df["Receita"].max()
```

```python
# Identificando a menor receita
df["Receita"].min()
```

```python
# Identificando o top 10 lojas em receita
df.nlargest(10, "Receita")
```

```python
# Identificando o Worst 10 lojas em receita
df.nsmallest(10, "Receita")
```

```python
# Agrupando as receitas por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
# Ordenando os valores das receitas agrupadas
df.sort_values("Receita", ascending=False).head(10)
```

```python
# Trabalhando com datas
```

```python
# Transformando a coluna de datas em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
# Verificando o tipo de dados de cada coluna
df.dtypes
```

```python
# Transformando a coluna data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# Agrupando por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
# Criando uma nova coluna com o Ano
df["Ano_venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
# Extraindo o mês e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
# retornando a data mais antiga da base de dados
df["Data"].min()
```

```python
# Retornado a diferença entre datas
df["diferenca_datas"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
# Criando a coluna de vendas de trimestre
df["trimestre_vendas"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
# Filtrando as vendas de 2019 - Mês Março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
vendas_marco_19.sample(20)
```
